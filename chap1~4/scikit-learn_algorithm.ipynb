```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd

iris = load_iris()

iris_data = iris.data #iris dataset 의 feature만을 numpy로 가지고 있음

iris_label = iris.target #iris dataset의 label(결정값)데이터를 numpy로 가지고 있음

print('iris target값', iris_label)
print('iris target명', iris.target_names)
```

```python
iris_df = pd.DataFrame(data = iris_data, columns = iris.feature_names)
iris_df
```

```python
iris_df['label']=iris.target
iris_df.head(3)
```

```python
#학습용, 테스트용 분리
#test_size 파라미터: 테스트데이터 비율
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size=0.2,random_state=11)
```

```python
#DecisionTreeClassifier객체생성하기
dt_clf = DecisionTreeClassifier(random_state = 11)

#학습수행
dt_clf.fit(X_train, y_train)
```

```python
#DTC객체의 predict()메소드에 테스트용 피처데이터세트 입력하면 학습된 모델 기반 테스트데이터 세트 대한 예측값 반환
pred = dt_clf.predict(X_test)
```

```python
#예측성능평가
from sklearn.metrics import accuracy_score
print('예측정확도:{0:.4f}'.format(accuracy_score(y_test,pred)))
```

교차검증

```python
# K-fold cross validation
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
import numpy as np

iris = load_iris()
features = iris.data
label = iris.target
dt_clf = DecisionTreeClassifier(random_state=156)

#5개의 fold set으로 분리하는 Kfold객체
#fold set별 정확도를 담을 리스트객체
kfold = KFold(n_splits=5)
cv_accuracy = []
print('붓꽃 데이터세트 크기: ',features.shape[0])
```

```python
#KFold객체에 .split()호출 ->학습용/검증용 데이터 분할 인덱스 반환
#실제 학습용/검증용 데이터 추출은 직접
n_iter = 0

for train_index, test_index in kfold.split(features):
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = label[train_index], label[test_index]
    #학습/예측
    dt_clf.fit(X_train, y_train)
    pred = dt_clf.predict(X_test)
    n_iter+=1
    #반복시마다 정확도 측정
    accuracy = np.round(accuracy_score(y_test, pred), 4)
    train_size = X_train.shape[0]
    test_size = X_test.shape[0]
    print('\n#{0} 교차 검증 정확도 :{1}, 학습 데이터 크기: {2}, 검증 데이터 크기: {3}'
          .format(n_iter, accuracy, train_size, test_size))
    cv_accuracy.append(accuracy)
    
#개별 iteration별 정확도 함쳐 평균 정확도 계산
print('\n## 평균 검증 정확도: ', np.mean(cv_accuracy))
```

Stratified K-fold

```python
import pandas as pd

iris = load_iris()

iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['label']=iris.target
iris_df['label'].value_counts()
```

```python
#k-fold의 문제
kfold = KFold(n_splits=3)
# kfold.split(X)는 폴드 세트를 3번 반복할 때마다 달라지는 학습/테스트 용 데이터 로우 인덱스 번호 반환. 
n_iter =0
for train_index, test_index  in kfold.split(iris_df):
    n_iter += 1
    label_train= iris_df['label'].iloc[train_index]
    label_test= iris_df['label'].iloc[test_index]
    print('## 교차 검증: {0}'.format(n_iter))
    print('학습 레이블 데이터 분포:\n', label_train.value_counts())
    print('검증 레이블 데이터 분포:\n', label_test.value_counts())
```

```python
#straified k-fold도 kfold와 비슷하게 사용하나
#split에서 피쳐데이터 뿐 아니라 레이블데이터세트도 필요
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=3)
n_iter=0
for train_index, test_index in skf.split(iris_df, iris_df['label']):
    n_iter += 1
    label_train = iris_df['label'].iloc[train_index]
    label_test = iris_df['label'].iloc[test_index]
    print('## 교차 검증: {0}'.format(n_iter))
    print('학습 레이블 데이터 분포:\n', label_train.value_counts())
    print('검증 레이블 데이터 분포:\n', label_test.value_counts())
```

```python
dt_clf = DecisionTreeClassifier(random_state=156)

skfold = StratifiedKFold(n_splits=3)
n_iter=0
cv_accuracy=[]

# 레이블 데이터 셋도 추가 입력
for train_index, test_index  in skfold.split(features, label):
    # split( )으로 반환된 인덱스를 이용하여 학습용, 검증용 테스트 데이터 추출
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = label[train_index], label[test_index]
    #학습 및 예측 
    dt_clf.fit(X_train , y_train)    
    pred = dt_clf.predict(X_test)

    # 반복 시 마다 정확도 측정 
    n_iter += 1
    accuracy = np.round(accuracy_score(y_test,pred), 4)
    train_size = X_train.shape[0]
    test_size = X_test.shape[0]
    print('\n#{0} 교차 검증 정확도 :{1}, 학습 데이터 크기: {2}, 검증 데이터 크기: {3}'
          .format(n_iter, accuracy, train_size, test_size))
    print('#{0} 검증 세트 인덱스:{1}'.format(n_iter,test_index))
    cv_accuracy.append(accuracy)
    
# 교차 검증별 정확도 및 평균 정확도 계산 
print('\n## 교차 검증별 정확도:', np.round(cv_accuracy, 4))
print('## 평균 검증 정확도:', np.mean(cv_accuracy))
```

GridSearchCV: 교차검증+최적하이퍼파라미터튜닝

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
#여기선 DTC의 주요 하이퍼 파라미터인 max_depth와 min_samples_split 값을 변화시키며 최적화 진행
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, test_size=0.2, random_state=121)
dtree = DecisionTreeClassifier()

parameters = {'max_depth':[1, 2, 3], 'min_samples_split':[2,3]}
```

```python
import pandas as pd

#학습데이터셋을 GridSearchCV의 fit 메소드에 인자로 입력하면 cv에 기술된 폴딩세트로 분할해 하이퍼 파라미터 순차적 변경하며 학습/평가
#그 결과를 cv_results_ 속성에 기록:GridSearchCV의 결과 set:딕셔너리형태
grid_dtree = GridSearchCV(dtree, param_grid=parameters,cv=3,refit=True)
#refit이 True이기 때문에 가장 좋은 파라미터 설정으로 재학습
grid_dtree.fit(X_train, y_train) #fit수행시 최적하이퍼파라미터값과평과결과값이 best_params_, best_score_속성에 기록

#cv_results DataFrame형태로 바꿔서 편하게
scores_df = pd.DataFrame(grid_dtree.cv_results_)
scores_df
```

```python
scores_df[['params', 'mean_test_score', 'rank_test_score', \
           'split0_test_score', 'split1_test_score', 'split2_test_score']]
```

```python
print('GridSearchCV 최적 파라미터:', grid_dtree.best_params_)
print('GridSearchCV 최고 정확도: {:.4f}'.format(grid_dtree.best_score_))
```

```python
#refit이 True였기 때문에 하이퍼 파라미터로 Estimator학습해 best_estimator_로 저장됨
# GridSearchCV의 refit으로 이미 학습이 된 estimator 반환
estimator = grid_dtree.best_estimator_


pred = estimator.predict(X_test)
print('테스트 데이터 세트 정확도: {0:.4f}'.format(accuracy_score(y_test,pred)))
```
